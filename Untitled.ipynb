```python
import os
import requests
from dotenv import load_dotenv
from smolagents import CodeAgent, InferenceClientModel, tool

# 1. Load the secret HF_TOKEN from your local .env file
load_dotenv()
hf_token = os.environ.get("HF_TOKEN")

if not hf_token:
    print("WARNING: HF_TOKEN not found. Did you create the .env file?")

# 2. Build the Tool
@tool
def fetch_real_recipe(dish_idea: str) -> str:
    """
    Fetches a real recipe, instructions, and ingredients from the live internet.
    
    Args:
        dish_idea: The name of the specific food or dish to search for (e.g., 'Lasagna', 'Chicken').
    """
    print(f"🤖 Alfred is searching the web for: {dish_idea}...")
    url = f"https://www.themealdb.com/api/json/v1/1/search.php?s={dish_idea}"
    
    try:
        response = requests.get(url).json()
        if response.get('meals'):
            meal = response['meals'][0] 
            name = meal['strMeal']
            category = meal['strCategory']
            instructions = meal['strInstructions'][:300] 
            return f"Found a {category} recipe for {name}! Here is how to start making it: {instructions}..."
        else:
            return f"I searched the database, but couldn't find a recipe for {dish_idea}."
    except Exception as e:
        return "The kitchen database is currently offline."

# 3. Initialize the Agent
agent = CodeAgent(
    tools=[fetch_real_recipe], 
    model=InferenceClientModel(api_key=hf_token), # Passing the token to the brain!
    max_steps = 5
)

# 4. Run it!
agent.run("Alfred, Mr. Wayne wants to eat something with spicy pasta. Find a real recipe.")
```

```python
from smolagents import CodeAgent, DuckDuckGoSearchTool, LiteLLMModel, GradioUI

# 1. The Brain: Your local Llama 3 model
model = LiteLLMModel(model_id="ollama/llama3")
options={
        "num_ctx": 8192  # This doubles the memory from 4096 to 8192
    }

# 2. The Tool: Free internet access
web_search = DuckDuckGoSearchTool()

# 3. The Agent: Assembling Alfred
agent = CodeAgent(
    tools=[web_search], 
    model=model,
    max_steps=2,
    name="Adarsh",
    description="The loyal, internet-connected butler of Wayne Manor."
)

# 4. The Interface: Launching the Web App
print("Booting up the Batcomputer UI...")
ui = GradioUI(agent)
ui.launch(share=False)
```
